```python
import os 
import re
```

```python
path = os.path.join('examples','hp_full_book.txt')
```

```python
path
```

```python
n = 0
with open(path, 'r') as file:
    word_list = file.read().split()
    print(len(word_list))
```

```python
word_list[:20]
```

```python
word_list[:20:2] #given the first 20 words, skip every other word
```

```python
word_list = [re.sub(r"^\s*", "", w) for w in word_list]
word_list = [re.sub(r"^\W*", "", w) for w in word_list]
word_list = [re.sub(r"[^\\p{IsPunctuation}]", "", w) for w in word_list]
word_list = [i for i in word_list if i != '']
word_list = [word.lower() for word in word_list]
```

```python
len(word_list)
```

```python
wrd_cnt = {}

for w in word_list:
    wrd_cnt[w] = wrd_cnt.get(w, 0) + 1
```

```python
def reverse_k(n):
    return -n[1]
```

```python
sorted(wrd_cnt.items(), key = reverse_k) #takes the tuples as the input to the function 
```
